```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python

# load the dataset, first dataset was created on csv file then loaded on collab python ide
dataset = pd.read_csv("/content/deep learning dataset of lab exam.csv")
# split into input (X) and output (y) variables
X = dataset.iloc[:, 0:3].values
y = dataset.iloc[:, 3].values
X
```

```python
dataset
```

```python
#standardisation od input data
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
X
```

```python
from keras import Sequential
from keras.layers import Dense
model = Sequential()
model.add(Dense(4, activation = 'relu',
                     kernel_initializer='random_normal',input_dim=3))
#Second Hidden Layer

model.add(Dense(4, activation = 'relu',
                     kernel_initializer = 'random_normal'))
#3rd hidden layer
model.add(Dense(6, activation = 'relu',
                     kernel_initializer = 'random_normal'))
#Output Layer
model.add(Dense(1,activation='sigmoid',
                     kernel_initializer='random_normal'))
```

```python
# compile the keras model
model.compile(optimizer="adam", loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X,y, batch_size=1, epochs=50)
```

```python
from mpl_toolkits import mplot3d
```

```python
figure = plt.figure()
axis_x = plt.axes(projection='3d')
```

```python

axis_x.scatter3D(dataset.iloc[:,0],dataset.iloc[:,1],dataset.iloc[:,2])
figure
```

```python
def f(x, y):
    return 4

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
axis_x.plot_surface(Z, X, Y, alpha=0.2)
axis_x.set_xlabel('x')
axis_x.set_ylabel('y')
axis_x.set_zlabel('z')
figure
```

```python
#predict the test data "t"
lst= {'x':[1,0],'y':[1,0]
        ,'z':[1,0]}

t=pd.DataFrame(lst)

y_predict=model.predict(t)
y_predict=(y_predict>0.5)
y_predict
```
