```
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

plt.rcParams['figure.facecolor'] = 'white'
```

```
def relu(x):
  return (np.maximum(x,0))

def sigmoid(x):
  return (1/(1+np.exp(-x)))

x = np.linspace(-10,10,200)
y = np.linspace(-2,2,200)
```

```
plt.plot(x,np.tanh(x))
plt.plot(x,np.zeros(x.shape), color = 'red')
plt.plot(np.zeros(x.shape),y, color = 'red')
plt.show()
```

```
plt.plot(x,sigmoid(x))
plt.plot(x,np.zeros(x.shape), color = 'red')
plt.plot(np.zeros(x.shape),y, color = 'red')
plt.show()
```

```
plt.plot(x,relu(x))
plt.plot(x,np.zeros(x.shape), color = 'red')
plt.plot(np.zeros(x.shape),y, color = 'red')
plt.show()
```

```

f1 = x**3 + x**2 + x + 2

plt.plot(x,f1)
plt.plot(x,np.zeros(x.shape), color = 'red')
plt.plot(np.zeros(x.shape),y, color = 'red')
```

```
f2 = x**2 + 5*x -10

plt.plot(x,f2)
plt.plot(x,np.zeros(x.shape), color = 'red')
plt.plot(np.zeros(x.shape),y, color = 'red')
```

```
f3 = 10*x + 30

plt.plot(x,f3)
plt.plot(x,np.zeros(x.shape), color = 'red')
plt.plot(np.zeros(x.shape),y, color = 'red')
```

```
o = relu(f1) - relu(f2) + relu(f3)

plt.plot(x,o)
plt.plot(x,np.zeros(x.shape), color = 'red')
plt.plot(np.zeros(x.shape),y, color = 'red')
plt.show()
```

```

o = -relu(f1) + relu(f2) + relu(f3)

plt.plot(x,o)
plt.plot(x,np.zeros(x.shape), color = 'red')
plt.plot(np.zeros(x.shape),y, color = 'red')
plt.show()
```

```
o = relu(f1) + relu(f2) + relu(f3)

plt.plot(x,o)
plt.show()
```

```
x = np.arange(-2,2,.2)


n1 = relu(-5*x -7.7)
n2 = relu(-1.2*x -1.3)
n3 = relu(1.2*x + 1)
n4 = relu(1.2*x -.2)
n5 = relu(2*x -1.1)
n6 = relu(5*x - 5)

z = -n1 -n2 -n3 +n4 +n5 +n6

plt.plot(x,z)
plt.plot(x,np.zeros(x.shape), color = 'red')
plt.plot(np.zeros(x.shape),x, color = 'red')
plt.show()
```
